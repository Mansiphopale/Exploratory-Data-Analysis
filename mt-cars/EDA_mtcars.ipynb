```python
mtcars = mtcars.drop_duplicates() 
```

```python
mtcars.shape
```

```python
mtcars = mtcars.drop(['Unnamed: 0', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16'], axis=1)
```

```python
mtcars.isnull().sum()
```

```python
mtcars["mpg"].fillna(mtcars["mpg"].mode()[0], inplace=True)
mtcars["cyl"].fillna(mtcars["cyl"].mode()[0], inplace=True)
mtcars["disp"].fillna(mtcars["disp"].mode()[0], inplace=True)

```

```python
# Fill missing values in numeric columns with their respective means
mtcars['hp'].fillna(mtcars['hp'].mean(), inplace=True)
mtcars['drat'].fillna(mtcars['drat'].mean(), inplace=True)
mtcars['wt'].fillna(mtcars['wt'].mean(), inplace=True)
mtcars['qsec'].fillna(mtcars['qsec'].mean(), inplace=True)
mtcars['vs'].fillna(mtcars['vs'].mean(), inplace=True)
mtcars['am'].fillna(mtcars['am'].mean(), inplace=True)
mtcars['gear'].fillna(mtcars['gear'].mean(), inplace=True)
mtcars['carb'].fillna(mtcars['carb'].mean(), inplace=True)



```

```python
mtcars.isnull().sum()
```

```python
mtcars.info()
```

```python
mtcars.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoders
le_mpg = LabelEncoder()
le_cyl = LabelEncoder()
le_disp = LabelEncoder()

```

```python
mtcars['mpg'] = le_mpg.fit_transform(mtcars['mpg'])
mtcars['cyl'] = le_cyl.fit_transform(mtcars['cyl'])
mtcars['disp'] = le_disp.fit_transform(mtcars['disp'])
```

```python
mtcars.info()
```

```python

# Correlation matrix and heatmap
#( A heatmap is a graphical representation of data in a matrix format,
#where individual values within the matrix are represented as colors. Heatmaps are commonly used to visualize the magnitude
#of a variable in relation to two factors, often displayed as a 2D grid or table.
#They are particularly useful for identifying patterns, correlations, or variations in data.)
correlation_matrix = mtcars.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
mtcars = pd.read_csv("C:/Users/Mansi/Downloads/mtcars (2) - mtcars.csv")
# Pairplot for selected columns
sns.pairplot(mtcars[['mpg', 'cyl', 'disp', 'hp', 'wt']])
plt.show()
```

```python

# Histogram of MPG
plt.figure(figsize=(8, 4))
sns.histplot(mtcars['mpg'], bins=10, kde=True)
plt.xlabel('Miles per Gallon (mpg)')
plt.title('Histogram of MPG')
plt.show()
```

```python

# Boxplots for categorical variables
plt.figure(figsize=(12, 6))
sns.boxplot(x='cyl', y='mpg', data=mtcars,palette='Set2')
plt.xlabel('Number of Cylinders')
plt.ylabel('Miles per Gallon (mpg)')
plt.title('Boxplot of MPG by Number of Cylinders')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(8, 4))

# Use a color palette for the plot
sns.countplot(x='am', data=mtcars, palette='Set2')  # Use a named palette, e.g., 'Set2'

# Add labels and title
plt.xlabel('Transmission Type (0 = Automatic, 1 = Manual)')
plt.ylabel('Count')
plt.title('Count of Cars by Transmission Type')

# Show the plot
plt.show()
```
