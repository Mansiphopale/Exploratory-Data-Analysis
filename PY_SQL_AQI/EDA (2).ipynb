Handling Outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot boxplot to detect outliersimport seaborn as sns
import matplotlib.pyplot as plt

# Visualize outliers using boxplots
plt.figure(figsize=(12,6))
sns.boxplot(data=df.select_dtypes(include=['number']))  # Only numeric columns
plt.xticks(rotation=45)
plt.show()
```

```python

# Remove outliers using IQR method
numeric_cols = df.select_dtypes(include=['number']).columns  # Select numeric columns
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

```

```python
# Filtering out outliers
df_cleaned = df[~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Original shape:", df.shape)
print("New shape after outlier removal:", df_cleaned.shape)

plt.figure(figsize=(12,6))
sns.boxplot(data=df.select_dtypes(include=['number']))  # Use only numeric data
```

```python

# Remove outliers using IQR method
numeric_cols = df.select_dtypes(include=['number']).columns  # Select numeric columns only
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

```

```python
# Apply IQR filtering only on numeric columns
df = df[~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

 Feature Engineering

```python
#  Create a Pollution Indicator
# Add a column indicating high pollution (Threshold: AQI > 100)
df['HighPollution'] = df['NO2(GT)'].apply(lambda x: 1 if x > 100 else 0)
```

```python
import pandas as pd

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Convert 'Time' to a proper format (if necessary)
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce').dt.time
```

```python
# Extract Year, Month, Day
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

# Extract Weekday (0 = Monday, 6 = Sunday)
df['Weekday'] = df['Date'].dt.weekday

# Identify if it's a weekend (1 for Yes, 0 for No)
df['IsWeekend'] = df['Weekday'].apply(lambda x: 1 if x >= 5 else 0)

# Extract Season (Winter, Spring, Summer, Autumn)
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df['Season'] = df['Month'].apply(get_season)
```

```python
print(df[['Date', 'Time', 'Year', 'Month', 'Day', 'Weekday', 'IsWeekend', 'Season']].head())
```

```python
#label encoding 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Season'] = le.fit_transform(df['Season'])
```

```python
#One-Hot Encoding (for non-ordinal data)
df = pd.get_dummies(df, columns=['Season'])
#This increases the number of features (which can impact model 
#complexity and training time)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr_matrix = df.corr()

# Display correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Check correlation of features with target variable (assuming 'PM 2.5' is the target)
target_var = "PT08.S5(O3)"  # Change to actual pollution variable if different

correlation_target = corr_matrix[target_var].abs().sort_values(ascending=False)
print(correlation_target)
```

ðŸ”¹ Actionable Insights:

Keep features with high correlation (e.g., > 0.4).
Drop highly correlated features (> 0.9) to remove redundancy.

```python
# Find pairs of highly correlated features
threshold = 0.9
high_corr_features = set()

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            colname = corr_matrix.columns[i]
            high_corr_features.add(colname)

# Drop redundant features
df_reduced = df.drop(columns=high_corr_features)
print("Dropped Features:", high_corr_features)
```

Feature Importance using Mutual Information

```python
from sklearn.feature_selection import mutual_info_regression

# Drop non-numeric columns
df_numeric = df.select_dtypes(include=['number']).dropna()

# Define target and features
X = df_numeric.drop(columns=[target_var])
y = df_numeric[target_var]

# Compute Mutual Information
mi_scores = mutual_info_regression(X, y)
mi_series = pd.Series(mi_scores, index=X.columns).sort_values(ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 5))
sns.barplot(x=mi_series, y=mi_series.index, palette="viridis")
plt.xlabel("Mutual Information Score")
plt.ylabel("Feature")
plt.title("Feature Importance (Mutual Information)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure target variable is correctly defined
target_var = "PT08.S5(O3)"  # Adjust based on actual target variable

# Select relevant features
selected_features = ["T", "RH", "PT08.S4(NO2)", "CO(GT)"]
df_final = df[selected_features + [target_var]].copy()  # Create a copy to avoid modification warnings

### ðŸ”¹ Pairplot for feature relationships
sns.pairplot(df_final)
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()

### ðŸ”¹ Scatter plots of features against the target variable
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Adjust grid layout for 4 features
fig.suptitle("Feature vs Target Variable", fontsize=16)

for i, feature in enumerate(selected_features):
    row, col = divmod(i, 2)  # Adjusting for a 2x2 subplot layout
    sns.scatterplot(data=df_final, x=df_final[feature], y=df_final[target_var], ax=axes[row, col])
    axes[row, col].set_title(f"{feature} vs {target_var}")

plt.tight_layout()
plt.show()

### ðŸ”¹ Heatmap to check correlations
plt.figure(figsize=(8, 6))
sns.heatmap(df_final.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Selected Features")
plt.show()
```
